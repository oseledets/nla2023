# Floating point: formula

$$f = (-1)^s 2^{(p-b)} \left( 1 + \frac{d_1}{2} + \frac{d_2}{2^2}  + \ldots + \frac{d_m}{2^m}\right),$$

where $s \in \{0, 1\}$ is the sign bit, $d_i \in \{0, 1\}$ is the $m$-bit mantissa, $p \in \mathbb{Z}; 0 \leq p \leq 2^e$, $e$ is the $e$-bit exponent, commonly defined as $2^e - 1$

Can be thought as a uniform $m$-bit grid between two sequential powers of $2$.

# Householder Reflections for matrices

$$
\begin{bmatrix}
* & * & *\\
* & * & * \\
* & * & *
\end{bmatrix}
\rightarrow
\begin{bmatrix}
* & * & * \\
0 & * & * \\
0 & * & *
\end{bmatrix}
\rightarrow
\begin{bmatrix}
* & * & * \\
0 & * & * \\
0 & 0 & *
\end{bmatrix}  $$

$$R = H_2 H_1 X$$

$$H2 =
\begin{bmatrix}
1 & 0 & 0\\
0 & * & * \\
0 & * & *
\end{bmatrix}
$$

$$ H_1^{-1} H_2^{-1} R = X $$

$$ H_1^T H_2^T R = X $$

$$ Q R = X $$

# Givens Rotations for vectors

$$\begin{pmatrix}
* \\ * \\ *
\end{pmatrix} \xrightarrow{\text{$G_1$}}
\begin{pmatrix}
* \\ * \\ 0
\end{pmatrix} \xrightarrow{\text{$G_2$}}
\begin{pmatrix}
* \\ 0 \\ 0
\end{pmatrix} $$

$$\begin{bmatrix}
\cos{\alpha} & -\sin{\alpha} \\
\sin{\alpha} & \cos{\alpha} \\
\end{bmatrix}$$

$$
G =
\begin{bmatrix}
1 & \dots & \dots & \dots & \dots & \dots & 0 \\
0 & \dots & \cos{\alpha} & \dots & -\sin{\alpha} & \dots & 0 \\
\dots & \dots & \dots & I & \dots & \dots & \dots \\
0 & \dots & \sin{\alpha} & \dots & \cos{\alpha} & \dots & 0 \\
0 & \dots & \dots & \dots & \dots & \dots & 1
\end{bmatrix}
$$

$$G[i,i] = \cos{\alpha}, G[i,j] = -\sin{\alpha}$$

$$G[j,i] = \sin{\alpha}, G[j,j] = \cos{\alpha}$$

## Derivation

$$\begin{bmatrix}
1 & 0 & 0 & 0 \\
0 & \cos{\alpha} & -\sin{\alpha} & 0 \\
0 & \sin{\alpha} & \cos{\alpha} & 0 \\
0 & 0 & 0 & 1
\end{bmatrix}
\begin{pmatrix}
x_1 \\ x_2 \\ x_3 \\ x_4
\end{pmatrix} =
\begin{pmatrix}
x_1 \\
x_2 \cos{\alpha} - x_3 \sin{\alpha} \\
x_2 \sin{\alpha} + x_3 \cos{\alpha} \\
x_4
\end{pmatrix}
$$ \
$$
\sin{\alpha} = - \frac{x_3}{\sqrt{x_2^2 + x_3^2}}, \cos{\alpha} = \frac{x_2}{\sqrt{x_2^2 + x_3^2}}
$$ \
$$
x_3' = 0 \\
x_2' = \frac{x_2^2 + x_3^2}{\sqrt{x_2^2 + x_3^2}}
$$

## In i, j notation:
$$
\sin{\alpha} = - \frac{x_j}{\sqrt{x_i^2 + x_j^2}}, \cos{\alpha} = \frac{x_i}{\sqrt{x_i^2 + x_j^2}}
$$ \
$$
x_j' = 0 \\
x_i' = \frac{x_i^2 + x_j^2}{\sqrt{x_i^2 + x_j^2}}
$$

# Givens Rotations for matrices

$$
\begin{bmatrix}
* & * & *\\
* & * & * \\
* & * & *
\end{bmatrix}
\xrightarrow{\text{$G_1$}}
\begin{bmatrix}
* & * & * \\
* & * & * \\
0 & * & *
\end{bmatrix}
\xrightarrow{\text{$G_2$}}
\begin{bmatrix}
* & * & * \\
0 & * & * \\
0 & * & *
\end{bmatrix}
\xrightarrow{\text{$G_3$}}
\begin{bmatrix}
* & * & * \\
0 & * & * \\
0 & 0 & *
\end{bmatrix} $$

$$R = G_3 G_2 G_1 X $$

$$G_1^{-1}G_2^{-1}G_1^{-1}R = X$$

$$G_1^{T}G_2^{T}G_1^{T}R = X$$
